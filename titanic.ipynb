```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report

#Data Prepartion
#Load the Dataset
df=pd.read_csv("titanic.csv")
df.describe()

#Handle missing values and categirical values
#Perform features enginering if needed

df_encode=pd.get_dummies(df,columns=['Sex'])
features=['Age','Pclass','Sex_male','Sex_female']

#choose a machine learning model
model=RandomForestClassifier()

#split the s=data into training and testing setd
X_train,X_test,y_train,y_test=train_test_split(df_encode[features],df_encode['Survived'],test_size=0.2,random_state=42)

#Train the model
model.fit(X_train,y_train)
y_pred=model.predict(X_test)

#Evalut the model
accuracy=accuracy_score(y_test,y_pred)
print('Accuracy:{:.2f}%'.format(accuracy*100))
print("classification Report:",classification_report(y_test,y_pred))
```
